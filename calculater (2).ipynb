```python
A = '******** HELLO********\n \t WELCOME TO CODERS REDAY !\n !!!ARE YOU READY TO SOLVE PROBLEM !!!\n PLESE SELECT THE OPTION \n \n \n 1.ADDITION\n 2.SUBSTRACTIN \n 3.MULTIPLICATION \n 4.DIVISION'
```

```python
from colorama import Fore, Back, Style
 
```

```python
class Calculator:
    def add(self,a,b):
        return a+b


    def subtract(self,a,b):
        return a-b


    def multiply(self,x,y):
        return x*y


    def divide(self,x,y):
        return x/y
    
# create calculater object
my_cl = Calculator()

print(Fore.RED + A)


while True:
    choice = input('enter choice 1/2/3/4/5 :  ')


    if choice in ('1','2','3','4','5'):

        if choice=='5':
            break

        num1 = float(input('enter a number: '))
        num2 = float(input('enter a second number: '))


        if choice =='1':
            print(num1 ,'+',num2 ,'\n\n YOUR ANSWER' ,my_cl.add(num1,num2))



        elif choice =='2':
                print(num1 ,'-',num2, '\n\n YOUR ANSWER' ,my_cl.subtract(num1,num2))

        elif choice =='3':
                print(num1 ,'X',num2, '\n\n YOUR ANSWER' ,my_cl.multiply(num1,num2))


        elif choice =='4':
                print(num1 ,'/',num2 ,'\n\n YOUR ANSWER' ,my_cl.divide(num1,num2))

        break

    else:
        print('invalild input')


```
